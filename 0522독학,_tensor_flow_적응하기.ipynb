판다스를 이용하여 데이터 가져오기

```python
import pandas as pd
data = pd.read_csv('/content/gpascore.csv')
print(data)
```

```python
data.isnull().sum() # gre행이 비어있다.
data = data.dropna() #빈칸의 행을 지워준다.
#data.fillna(100) #빈칸에 값을 넣어줌 100
data.isnull().sum()
```

```python
print(data['gpa'])
```

```python
y데이터 = data['admit'].values # admit행의 모든값을 리스트형태로 저장
print(y데이터)
```

```python
print(x데이터) # 데이터가 다 들어간것을 확인할 수 있음.
```

# 혼자 다른 데이터로 복습해보기.

이번에는 데이터 가공먼저 해볼것이야.

```python
import pandas as pd # 데이터를 가져오기위한 패키지
import numpy as np # 가져온 데이터를 딥러닝 모델에 넣기위한 모델
data = pd.read_csv('/content/Fish1.csv')
print(data)
```

```python
data.isnull().sum() # 전처리할 데이터 1로 모두 수정
ydata = data.fillna(1)
ydata = ydata['Species'].values #.values #Species 안에 Bream만 뽑고 싶은데 모르겠다 ㅠ,,ㅠ _y데이터 확보 완료
print(ydata) 
```

```python
xdata=[]
for i,rows in data.iterrows():
  xdata.append([rows['Weight'],rows['Length1'],rows['Height'],rows['Width']]) #x데이터 확보 완료
```

```python
print(ydata)
```

모델을 생성해보자 !

```python
import tensorflow as tf

models = tf.keras.models.Sequential([
    tf.keras.layers.Dense(64,activation='relu'),
    tf.keras.layers.Dense(64,activation='tanh'),
    tf.keras.layers.Dense(1,activation='sigmoid')                        
])
# 레이어가 3개인 모델 생성
```

```python
models.compile(optimizer='adam',loss='binary_crossentropy',metrics='accuracy')
```

```python
models.fit(np.array(xdata),np.array(ydata),epochs=100)
```

```python
# 오류 - `logits` and `labels` must have the same shape, received ((None, 64) vs (None, 1)) 해결해보기
print(np.array(xdata).shape)
print(np.array(ydata).shape)
# 모델 생성할때 마지막 sigmoid라인을 64 노드 생성으로 했어 ㅡㅡ 아 ..
```
